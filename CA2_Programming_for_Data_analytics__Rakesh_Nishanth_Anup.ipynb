**Importing the required python packages or libraries in order to perform web scrapping**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

**Webscrapping and creating of dataset code goes here**

```python
#lists to store data
Product_names=[]
Prices=[]
Description=[]
Reviews=[]
No_of_Ratings=[]

for i in range(2,70):
    url="https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&page="+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.text,'lxml')
    box=soup.find("div",class_="_1YokD2 _3Mn1Gg")
    names=box.find_all("div",class_="_4rR01T")
    for n in names:
        name=n.text
        try:
          Product_names.append(name)
        except:
            Product_names.append(None)
    #print(Product_names)
    prices=box.find_all("div",class_="_30jeq3 _1_WHN1") 
    for p in prices:
        price=p.text.split('₹')[1]
        try:
          Prices.append(price)
        except:
            Prices.append(None)
    #print(Prices)
    descp=box.find_all("div",class_="fMghEO")
    for d in descp:
        des=d.text
        try:
          Description.append(des)
        except:
            Description.append(None)
    #print(Description)
    revs=box.find_all("div",class_="_3LWZlK")
    for r in revs:
        review=r.text
        try:
            Reviews.append(review)
        except:
            Reviews.append(None)
    #print(Reviews)
    noofRatings=box.find_all("span",class_="_2_R_DZ")
    for r in noofRatings:
        rating=r.text.split(' ')[0]
        try:
            No_of_Ratings.append(rating)
        except:
            No_of_Ratings.append(None)
    #print(No_of_Ratings)
    
productData = {"Product Name":Product_names,"Prices":Prices,"Description":Description,"Reviews":Reviews,"No. of Ratings": No_of_Ratings}
df=pd.DataFrame.from_dict(productData,orient="index")
df=df.transpose()
#df=pd.DataFrame({"Product Name":Product_names,"Prices":Prices,"Description":Description,"Reviews":Reviews,"No of Ratings": No_of_Ratings})
print(df)
df.head()

#while True:
   # np=soup.find("a",class_="_1LKTO3").get("href")
    #print(np)
    #cnp="https://www.flipkart.com"+np
    #print(cnp)
    #url=cnp
    #page=requests.get(url)
    #soup=BeautifulSoup(page.text,'html.parser')
```

The Python script listed above web scrapes the Flipkart website to gather information about cell phones priced under 50,000 rupees. A for loop is used to extract the data from several pages, and for each page, the script gathers the product name, price, description, customer reviews, and number of ratings for each mobile phone.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Concatenate all the descriptions into a single string
text = ' '.join(df['Description'].values)

# Tokenize the text into individual words
tokens = word_tokenize(text)

# Remove stop words and punctuation
stopwords = nltk.corpus.stopwords.words('english')
filtered_tokens = [word.lower() for word in tokens if word.lower() not in stopwords and word.isalpha()]

# Compute the frequency distribution of the tokens
fdist = FreqDist(filtered_tokens)

# Print the 15 most common features
print(fdist.most_common(15))

```

**QUESTION8:** *What is the distribution of the total number of ratings for different phone categories (budget, smart, premium) based on their review ratings (high, medium, low)?*

```python
# Create a pivot table to aggregate the number of ratings by category and review
ratings = pd.pivot_table(df, values='No. of Ratings', index='Label_prices', columns='Label_Reviews')

# Reset the index to create separate columns for each review category
ratings = ratings.reset_index()

# Melt the ratings DataFrame to create a long-form DataFrame for the stacked bar chart
ratings_melt = pd.melt(ratings, id_vars='Label_prices', var_name='Label_Reviews', value_name='No. of Ratings')

# Create a stacked bar chart
fig = px.bar(ratings_melt, x='Label_prices', y='No. of Ratings', color='Label_Reviews', barmode='stack')

# Set the title and axis labels
fig.update_layout(title='Number of Ratings by Category and Review', xaxis_title='Category',
                  yaxis_title='Total Number of Ratings')

# Show the plot
fig.show()
```

We intended to address this problem by producing a histogram; we see a peak in the frequency of ratings in the premium Phones category, indicating that there are more people who favor premium phones. The histogram also reveals that fewer users ranked the mobile phones as medium, implying that medium ratings are less common in the sample.

**QUESTION9:** How do the average prices of budget, smart, and premium phones vary based on their review ratings (high, medium, low)?

```python
# Calculate the average prices by category and review
avg_prices = df.groupby(['Label_prices', 'Label_Reviews']).mean()[['Prices']]

# Reset the index to create separate columns for each review category
avg_prices = avg_prices.reset_index()

# Create a comparison bar chart
fig = px.bar(avg_prices, x='Label_prices', y='Prices', color='Label_Reviews', barmode='group')

# Set the title and axis labels
fig.update_layout(title='Average Prices by Category and Review', xaxis_title='Category',
                  yaxis_title='Average Price')

# Show the plot
fig.show()
```

Based on the ratings, we are attempting to determine the average price for each category. The average cost of a Premium phone with a high or medium rating is 33,387 INR and 20,062 INR, respectively. The average cost of a budget phone with high and medium ratings is 13,736 INR and 14,732 INR, respectively. The average price of a smart phone with a high or medium rating is 19,658 INR and 15,624 INR, respectively.
